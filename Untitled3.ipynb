# **正規化について**
この章からは、正規化についてまとめたい。
<br>リレーショナルデータベースの設計において、データの重複をなくし、整合的にデータを取り扱えるように設計することをデータベースの正規化という。正規化を行っておくと、データを追加したり削除したりする場合にデータの不整合が起こるのを防ぎ、メンテナンス効率を高めることができる。
<br>正規化には第1～第5正規形があるが、ここではデータベースを設計する場合に一般的に用いられる第1～第3正規形までを取り扱う。正規化はデータの整合性を保ちやすくなる半面、テーブルの数が増える事で、結合操作がより必要になり検索SQLのパフォーマンスが劣化すると言ったデメリットがある。第1、第2…と進むごとに検索パフォーマンスは低くなり、データの整合性は高くなる。一般的には第3正規形までで十分と言われている。
## <br>第1正規形
<br>第1正規形の定義は「非正規形のテーブルから繰り返しを取り除いたもの」である。例えばとある塾の個別データとして以下のような表があったと仮定する。この表は、非正規形である。
<br>![非正規形](https://drive.google.com/uc?id=1v51n2B5WHmpkz1kPNlI0rp_qofQ5o7MG)
<br>この表は指導曜日がスカラ値になっていないためである。指導曜日が複数ある場合、上のように表記したくなるが、リレーショナルデータベースでは規則違反になる。このとき以下のように変形したものが第1正規形である。
<br>![第1正規形](https://drive.google.com/uc?id=1d9l_pzYrPLaBlh-Np5vM2Na0OMEvrB94)
<br>指導曜日の数だけ行数を増やした。第1正規形で必要なことは、繰り返し部分に一つ一つの独立したレコードを作り、横方向の長さを統一することである。
<br>このようにスカラ値にしなくてはならない理由は主キーが各列の値を一意に特定できないためである。ここで関数従属性という概念を取り上げる。関数従属性とは、リレーショナルデータベースの表で、2つの属性の間で片方の値を決定するともう一方の値が一意に決まるという性質のことである。非正規形の表のままであれば、{生徒ID}→{指導曜日}とい従属関係は成立しない。一方で、第一正規形の表は、{生徒ID}→{指導曜日} という従属関係が成り立っている。
## <br>第2正規形
<br>第2正規形の定義は、「部分関数従属している列が切り出されたもの」である。再度以下のような表を考える。
<br>![第1正規形](https://drive.google.com/uc?id=1d9l_pzYrPLaBlh-Np5vM2Na0OMEvrB94)
<br>これは先ほど正規化を施し、第1正規形となっている。しかし、この表において所属する生徒はいないが、とある学校の情報を入力したい場合、主キーの一部に生徒IDが含まれているため、生徒IDがNULL値になり登録ができない。
<br>このテーブルの主キーは「学校コード、生徒ID」である。この表を見ると、生徒名、指導曜日、教科、講師ID、講師名は学校コード、生徒IDに従属しているのに対し、学校名は学校コードのみに従属している。このような関係を「部分関数従属」という。これに対し主キーを構成する全ての列に従属性がある場合を「完全関数従属性」と言う。つまり、第1正規形の表の部分関数従属している列を切り出し、従属関係を分離することで第2正規形となる。以下は、上の表を第2正規形と変形している。
<br>![第2正規形](https://drive.google.com/uc?id=10a9_LRoJIiDtLiJtyp3CBweTnlwGd0nm)
<br>この表であれば、生徒のいない学校であっても情報を登録することができる。
## <br>第3正規形
<br>先で第2正規形まで変形したが、まだ不具合が生じる。例えば、まだ特定の生徒を担当していない講師「後藤」がいたとする。しかし、後藤の情報を入力すると、学校コード、生徒IDといった主キーがNULL値となり、登録ができない。このような不具合が起こるのは、先程の第2正規形の表に、隠れた従属関係が潜んでいるからである。

# Reduxについて
この章からはReduxについてまとめたい。
<br>Reduxとは「Action」と呼ばれるイベントを使用して、アプリケーションのState（状態）を管理し、更新するためのフレームワークである。 アプリケーションの多くの部分で必要なStateを、グローバルなStateとして一元管理するのに役立つ。Actionとはユーザーが何をしたいかという情報を持つオブジェクトのこと。ReduXはReact、jQuery、Angular、Vueといったライブラリと併用できる。

## <br>Reduxの原則について
### <br>原則①　Single source of truth
<br>アプリケーション全体の状態(State)はツリーの形で１つのオブジェクトで作られ、１つのストアに保存される。つまり、Storeを複数作成することはできない。またこの原則により、Stateが保存しやすいので、ユニバーサルアプリケーションがつくりやすい、Stateが1つであるため、デバッグしやすい、開発しやすいという特徴がある。
### <br>原則②　State Is read Only
<br>この原則は、状態を変更する手段は変更内容を持ったActionオブジェクトを実行するのみであることを示している。つまりビューやコールバックが状態を直接変更させることはできない。またActionはオブジェクトなので、保存可能であり、テストしやすいということも示している。
### <br>原則③　Mutations are written as pure functions
<br>和訳すると、「変更はすべて純粋関数で行われる」となる。粋関数とは引数にどのような値を与えても必ず同じ値を戻り値として返すということ。これを「Reducerは副作用がない」という呼び方をする。ReduxのReducerは純粋関数の思想が取り入れられているが、これには関数の再利用性や単体テストがしやすいというメリットがある。
<br>
<br>以上の3原則により、データの流れが一方向になり複雑にならず、プログラムの流れが追いやすくなる。

## <br>データフロー
<br>以下の図はReduxの概念を図式化したものである。この図を基に、今回はREduxの主要概念と用語についてまとめたい。
<br>![Redux概念図](https://drive.google.com/uc?id=1gO27U05MSsDIudIAIME0LkR5zQslAIia)

### <br>①Actionを生成
<br>Actionとは、先述のとおり
ユーザーが何をしたいかという情報を持つオブジェクト、つまりアプリケーション上でユーザーが行った「ボタンを押す動作」などと考えてよい。まずは、User Input　→　Viwe(compornent)　→　Actioncreater という流れでViewからActioncreaterにrequestが送られる。つまり、Compornent上からActionの作成依頼をActioncreaterに飛ばし、ActioncreaterからView(Compornent)にActionを返すといった流れだ。
<br>![action](https://drive.google.com/uc?id=10QrA4_EvqpuyspGudvFztjQ5emqPB91b)

###<br>②Actionをdispatchする
<br>続いて、先ほど作成したActionをStoreにdispatch(送信)する。Storeとは、Stateの状態を保持している場所のことで、dispatchすることでActionをStoreに送ることができる。Redux StoreのDispatchと呼ばれるメソッドは、状態を更新する唯一の方法で、Actionオブジェクトを引数にとる。Storeにイベントが発生したことを知らせる役割を担っている。
<br>![dispatch](https://drive.google.com/uc?id=1BQlZjqpNcTr6dBtTzIne17TfR1ZlaN0S)

###<br>ReducerがStateを更新
<br>Reducerとは、Actionを元にStateを更新するメソッドのこと。DispatchされたActionと現在保持しているStateが、Store内のReducerに渡され、新しいStateを作成する。つまりDispatchによりStoreにイベントが発生したことが知らされると、Reducerは現状を保持していたStateとActionを受け取り、新たな状態を作り出す。この状態を監視するためのシステムがReduxである。
<br>
<br>最後に、Reducerによって新たに作成されたStateをもとに、UIが更新される。
<br>
<br>
以上がReduxのデータフローの概要である。まとめると、
<br>**①ActionCreatorsによってActionを生成する**
<br>**②Actionをdispatchする**
<br>**③ReducerによってStore内のStateを更新する**
<br>**④ReactとReduxの連携しStore内のStateをComponentで参照する **
<br>となる。

<br>（※「フルスタックなフレームワーク」は、ソフトウェア開発のための包括的なフレームワークで、通常、フロントエンド（クライアントサイド）とバックエンド（サーバサイド）の両方の開発に関するツールや機能を提供する。これにより、開発者はアプリケーションの全体的な構築を同じフレームワーク内で行うことができる。）
<br>
<br>　そのため、新しいReactプロジェクトを開始するための、例えばRuby on Railsにおける「rails new」のような単一のコマンドは存在しない。
<br>　しかし、Reactで中規模以上のアプリを開発する場合、いくつかのステップが必要になる。これには、最新のJavaScriptやJSXを、古いブラウザでも実行可能なコードに変換するためのコンパイラ、JavaScriptやCSSのファイルを結合し最小化するためのバンドラなどが含まれる。これらのツールやライブラリを組み合わせて使うためには、それらが協力して動作するように複雑な設定が必要である。
<br>
<br>(※バンドラ・・・ウェブ開発において異なるソースコードやリソースを一つにまとめ、最適化するツールのことを指す。これにより、ウェブアプリケーションのパフォーマンスを向上させることができる。)
<br>(※コンパイラ・・・プログラムを一種の言語から別の言語に変換するソフトウェアツール。通常、高水準言語（人が理解しやすい形式）で書かれたソースコードを、低水準言語や機械語に変換する。)
<br>
<br>　まとめると、Reactは最初からすべてが揃ったフレームワークではないため、アプリケーションを構築するにはいくつかのツールや設定が必要になるということだ。
<br>　アプリ開発において前述した設定は大変である。その為、React のアプリケーションプロジェクトを作成・運用するためのツールが公式やサードパーティから提供されてる。その中の一つが Viteである。

##JSX について
以下、りあクト! TypeScript で始めるつらくない React 開発 第 4 版【2React基礎編】の 5.JSXでUIを表現する　を参照し、JSXについてまとめたい。
<br>
<br>
従来、開発者はマークアップとロジックを別々のファイルに分け、構造にはHTML、スタイリングにはCSSを使用し、インタラクションやデータ操作にはJavaScriptを記述してきた。JSX（JavaScript XML）は、JavaScriptの構文拡張で、JavaScriptファイル内にHTMLのようなコードを記述できるようにするもの。Meta（旧Facebook）によって開発された。JSXの構文はHTMLに似ており、開閉タグ、属性、入れ子要素などがある。
<br>
###<font color="red">「JSX は syntax sugarである」ということについて</font>
<br>JSX は、JavaScript XML の略で、React で使用される syntax sugar である。syntax sugar とは特定の機能や処理をより簡潔に記述するための構文のことを指し、JSX は、JavaScript で HTML ライクなコードを書くための構文ということになる。つまり、JSX を使用すると、JavaScript 内で XML のような構文を使用できるが、これは構文拡張であり、JavaScript の一部ではないことを表わしている。
<br>
<br>JSXは、ブラウザで実行される前に、通常のJavaScriptに変換される。この変換は、トランスパイラと呼ばれるツールを使って行われる。JSXの最も一般的なトランスパイラはBabelである。Babelは、JSXのコードを一連の関数呼び出しに変換する。これらの関数呼び出しは、JSXで書かれたHTMLのようなコードと同等になる。そして、ブラウザは、結果として得られたJavaScriptコードを実行することができる。
<br>例えば次のJSXコードを見る。
<br>`const element = <h1>Hello, world!</h1>;`
<br>これは次のようなJavaScriptコードに変換される。
<br>`const element = React.createElement("h1", null, "Hello, world!");`
<br>この変換により、開発者はJavaScriptの性能と柔軟性を生かしながら、慣れ親しんだ読みやすい構文でコードを書くことができるのだ。
<br>
<br>
##Reactがロジックとデザインを混合させる理由
<br>ほとんどのサーバサイドWebアプリケーションシステムは、アーキテクチャにMVCを用いている。
<br>ソフトウェア工学における「アーキテクチャ」や「デザインパターン」は、関心の分離を行うための重要な概念である。関心の分離とは、ソフトウェアシステムを設計する際に、異なる問題や課題を明確に分離して管理することで、システム全体の理解、保守性、拡張性を向上させることを指す。
<br>例えば前述したMVCモデルがどのような関心で要素を区分しているかといえば、アプリケーション全体で、技術的な役割によって3つの要素を区分していると言える。
<br>しかしながら、MVCモデルはフロントエンド開発においてはうまく機能しないという考え方がされるようになってきた。その理由は、Reactにおいては、関心の分離の単位が技術的な役割ではなく、アプリケーションにおける機能一つ一つとなっているからだ。
<br>MVCモデルを採用する多くでは、ページの表示に必要な処理をController が終わらせて、Viewでページ全体をまとめてレンダリングする。しかしReactではページを構成する部品であるコンポーネント自身が、必要な処理を自律的に行い、レンダリングも非同期的に並列で行われるため、これを分けようとするのが構造的に無理だと言える。
<br>これがReactが一つ一つのパーツの中に、ロジックとデザインを混合させる理由になっている。
<br>そして、コンポーネントのマークアップとロジックを1つのファイルに記述するために、JSXはReactに不可欠な要素となるのである。
<br>

#コンポーネントの基本
##コンポーネントのメンタルモデル
<br>上記までで、React がコンポーネントという独立した再利用可能なパーツを組み合せることによってアプリケーションを作るもの。つまり、コンポーネントベース・アーキテクチャを採用している技術だと確認してきた。
<br>ちょうどJSXで記述したように、React はメモリ上にReactElement オブジェクトのツリーを持っており、それを HTMLElement による DOM ツリーに還元し、ブラウザに出力することでアプリケーションとして動作する。その中では、React Elements というものが、コンポーネントを任意の props でコールするための実行リンクのようなものだと記述した。
<br>　では、React ではコンポーネントをどう考えるか。この答えとして、広く言われることだが、**プログラムの関数のようなもの**と考えるのが主流である。関数に倣って説明するならば、**props を引数として受け取り、戻り値として React Elementsを返す**という具合である。
<br>
<br>Reactにおけるコンポーネントを関数のようなものと記述したが、では違いはどこにあるのだろうか。コンポーネントが通常のプログラミングにおける関数とちがうのは、**個々に「状態」を持つことができる**、という点にある。
<br>　関数として定義されたReactコンポーネントは、内部に状態（state）を持つことができる。状態は、コンポーネント内部で管理され、必要に応じて変更することができる。これにより、コンポーネントは時間の経過とともに変化するデータを表現することができる。状態は、useStateフックを使用して定義され、setState関数を介して更新される。
<br>
##コンポーネントに Props を受け渡す
<br>これも先に説明しているが、props とはコンポーネントにとっての関数に対する引数にあたるものである。また、JSX におけるマウント時のタグの中では、そのタグの属性値として記述される。そして JSX が生成する React Elements を通してコールされたコンポーネント側には、それが { 属性名: 属性値 } の形式のオブジェクトとして渡される。そのコンポーネントが関数で実装されている場合、props のオブジェクトを第 1 引数として受け取る。
<br>上記の説明として以下のように定義されたコンポーネントをみる。




```python
function Greeting(props) {
  return <div>Hello, {props.name}!</div>;
}
```

この例では、propsはオブジェクトとしてGreetingコンポーネントに渡され、その中のnameプロパティを使ってコンポーネント内で表示している。
<br>
<br>
